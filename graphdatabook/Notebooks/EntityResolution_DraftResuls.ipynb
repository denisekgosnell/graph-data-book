```python
from gremlin_python.process.traversal import Scope, Column, Order, Direction
```

```python
top_rank = {}
for email_id, time in time_splits.items():
    
    answer = g.V().has("email_address", "email_id", email_id). \
        bothE().has("datetime", P.lt(time)).inV(). \
        bothE().has("datetime", P.gt(time)).inV(). \
        groupCount().by("email_id"). \
        order(Scope.local).by(Column.values, Order.desc). \
        unfold(). \
        limit(1).toList()
    top_rank[email_id] = answer
```

```python
thresholds = list(set(sorted(normalized.values())))
```

```python
results = {}
for thresh in thresholds:
    total = 0
    correct = 0
    threshold = int(thresh)
    for vertex, answer in top_rank.items():
        if int(normalized.get(vertex)) >= threshold:
            if len(answer) > 0:
                total +=1
                answer_v = answer[0].keys()
                for key in answer_v:
                    if key == vertex:
                        correct +=1
    if total > 0:
        results[thresh] = float(correct/total)
        
```
